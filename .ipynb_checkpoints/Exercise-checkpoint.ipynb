```python
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

# Import Data

```python
# Get the directory: 
d = os.getcwd() + '\data\\'
```

```python
# We will only need calendar.csv and listings.csv
```

```python
# Calendar
calendar_df = pd.read_csv(d + 'calendar.csv')
calendar_df.head()
```

```python
# Listings
listings_df = pd.read_csv(d + 'listings.csv')
listings_df.head()
```

# EDA

## Calendar

### Clean

```python
# Info
calendar_df.info()
```

```python
# Remove the $ from 'price' column and convert to float
calendar_df['price'] = calendar_df['price'].apply(lambda x: x.replace('$', '').replace(',', '') if type(x) != float else x)
calendar_df['price'] = calendar_df['price'].astype('float32')
```

```python
# Change the 'date' field to datetime type
calendar_df['date'] = pd.to_datetime(calendar_df['date'])
```

### Insights

```python
# Describe
calendar_df.describe()
```

```python
# Filter out available data
avai_df = calendar_df[calendar_df['available'] == 't']
avai_df
```

```python
# Sum by day

date_price_df = avai_df.groupby('date').price.sum()
date_price_df
```

```python
# Booked

listing_num = calendar_df['listing_id'].nunique()

booked_df = calendar_df[calendar_df['available'] == 't'].groupby('date').listing_id.nunique()
booked_df = booked_df.reset_index()
booked_df['booked'] = round(booked_df['listing_id']/listing_num * 100,2)
booked_df
```

```python
# Plot

fig, ax = plt.subplots(1, 1, figsize = (24, 6))

x = date_price_df.index
y = date_price_df.values
y2 = booked_df['booked']

ax.bar(x, y, color = 'b', alpha=0.3)
ax.xaxis_date()

ax2 = ax.twinx()
ax2.plot(x, y2)
ax2.set_ylim(0,100)


plt.show()
```

```python
# Drill down to 3 months: from 2016-09-01 to 2016-12-31

m_price_df = date_price_df['2016-09-01' : '2016-12-31']
```

```python
# Plot

fig, ax = plt.subplots(1, 1, figsize = (24, 6))

x = m_price_df.index
y = m_price_df.values

# Highlight friday and saturday
colors = ['r' if i.day_name() in ['Friday', 'Saturday'] else 'b' for i in m_price_df.index  ]

ax.bar(x, y, color=colors)
ax.xaxis_date()

# Start at 300000
ax.set_ylim(300000)

plt.show()
```
